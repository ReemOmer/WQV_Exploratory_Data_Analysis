```python
# Importing libraries and resizing the plot
from MLModels1 import MLModels
import matplotlib.pyplot as plt
```

```python
features = ['106604_00010','106612_00003','106616_00400','106618_32318']

for i in features:
    # Cleaning and spliting the data into training/testing
    x_train, y_train, x_test, y_test,date_measurements = MLModels.data_clean_split(i)
    # Creating a list to add the model's results to
    y_predict = []
    # Adding MultiLayer Perceptron to the list
    y_predict.append(MLModels.MLP_model(x_train,y_train,x_test,y_test))
    # Adding Linear regression results to the list
    y_predict.append(MLModels.LR_model(x_train,y_train,x_test))
    # Adding Logistic regression results
    y_predict.append(MLModels.MOR_model(x_train,y_train,x_test))
    # Adding Decision Tree Regressor
    y_predict.append(MLModels.DTR_model(x_train,y_train,x_test))
    # Adding Deep Neural Networks
    y_predict.append(MLModels.DNN_model(x_train,y_train,x_test,MLModels.baseline_model)) 
    # Adding Autoregression
    AR_test,AR_predict=MLModels.AR_model(date_measurements)
    y_predict.append(AR_predict)
    # Plot actual vs predicted values for the models
    plt.figure(figsize=(16,10))
    MLModels.plot_model(y_test,AR_test,y_predict)
    # Finding the accuracy of each model
    MLModels.acc_model(y_test,AR_test,y_predict)
```
